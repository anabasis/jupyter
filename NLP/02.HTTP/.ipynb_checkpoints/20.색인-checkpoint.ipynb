```python
def filecontent(file) :
    with open(file, encoding="utf-8") as fp :
        content = fp.read()
    return content
```

```python
#원시어절, 구두점, 불용어, 길이, 빈도, 정규식, 품사분석(형태소), ngram
from nltk.tokenize import word_tokenize
from konlpy.tag import Komoran

ma = Komoran()

for file in fileids("./news/") :
    with open(file, encoding="utf-8") as fp :
        content = fp.read() 
    tokens1 = content.split() # 원시어절
    #tokens2 = [word_tokenize(token) for token in tokens1]
    tokens2 = word_tokenize(content) # 구두점 분리 => 원시어절
    #tokens3 = [ ma.pos(token) for token in tokens2 ]  # => ("형태소","품사") 
    tokens3 = [ _ for token in tokens2 for _ in ma.pos(token)] # 형태소, 품사
    # token1 + token2 + token3 => Lexicon(Controlleds vocab)
    tokens4 = [ token[0] for token in tokens3 ]  # 형태소
    tokens5 = [ token[0] for token in tokens3 if token[1].startswith("N") ] #명사
    tokens6 = [ _ for token in tokens4 for _ in ngram(token)] # ngram
    
    #print(len(content.split()))
    print(len(tokens1))
    print(len(tokens1+tokens2+tokens3))
    print(len(set(tokens1+tokens2+tokens3 + tokens4+tokens5+tokens6)))
    
    break
```

```python
def ngram(term, n=2) :
    return [term[i:i+n] for i in range(len(term) - n + 1)]
    
```

```python
from collections import defaultdict

corpus = filecontent(fileids("./news/")[-2])
indexTerm = defaultdict(int)
indexTerm1 = defaultdict(int)
indexTerm2 = defaultdict(int)
indexTerm3 = defaultdict(int)
indexTerm4 = defaultdict(int)
indexTerm5 = defaultdict(int)

for _ in ["punc", "stopword", "email", "domain",  "whitespace", "nonkorea"] :
    corpus = pattern[_].sub(" ", corpus)
#content

# 원시어절
for term in word_tokenize(content) :
    indexTerm1[term] += 1  # 원시 어절
    indexTerm[term] += 1  

# 토플    
for _ in indexTerm1 :
    for t in ma.pos(_) :
        indexTerm2[t] += 1  # 원시형태소 + 품사
        indexTerm[t[0]] += 1  
        if len(t[0]) > 1 :   # 음절 길이로 정규화
            indexTerm3[t[0]] += 1  # 원시형태소
            indexTerm[t[0]] += 1  
        if t[1].startswith("N"):
            indexTerm4[t[0]] += 1  # 명사
            indexTerm[t[0]] += 1  
        for n in ngram(t[0]) :   # Bigram
            indexTerm5[n] += 1  # 
            indexTerm[n] += 1  

len(indexTerm1), len(indexTerm2), len(indexTerm3), len(indexTerm4), len(indexTerm), \
min(indexTerm5.values()),max(indexTerm5.values())
# => Lexicon(Controlled Vocaburary)
# => Classification(빈도, 상위 %)
# => Indexing
indexTerm
```

```python
# Lexicon
DTM = defaultdict(lambda:defaultdict(int))
for docName, docContent in collection :
    for term in word_tokenize(docContent):
        for morpheme in ma.pos(term):
            if len(morpheme[0]) > 1 and morpheme[1].startswith("N") :
                DTM[docName][term] += 1
```

```python
TDM = defaultdict(lambda:defaultdict(int))
for idx, termList in DTM.items():
    for term, freq in termList.items():
        TDM[term][idx] = freq
```
