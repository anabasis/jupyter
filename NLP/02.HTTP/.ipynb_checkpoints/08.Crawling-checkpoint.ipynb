```python
import requests
# 200 => OK
# 400 => My Fault
# 500 => You Fault => Traffic 과도하게 증가
header={"User-Agent":"Mozilla/5.0 (Windows NT 10.0; WOW64; Trident/7.0; rv:11.0) like Gecko"}

def download(url, params ={}, retries = 3) :
    resp = None
    try :
        resp = requests.get(url, params = params, headers = header)
        resp.raise_for_status()
    except requests.exceptions.HTTPError as e :
        if 500 <= e.response.status_code < 600 and retries > 0:
            print(retries)
            resp = download(url, params, retries - 1)
        else :
            print(e.response.status_code)
            print(e.response.reason)
            print(e.response.headers)
    
    return resp
```

```python
def urlExtractor(seed) :
    html = download(seed)
    dom = BeautifulSoup(html.text, "lxml")
    links = [_["href"] for _ in dom.select("a")]
    unseen = list()
    
    for link in links :
        if link[0] == "/" :
            newLink = requests.compat.urljoin(seed,link)
            if newLink not in unseen :
                unseen.append(newLink)
            else
                print("Skipped2", link)
        elif link.startswith("http") :
            unseen.append(link)
        else :
            print("Skipped1", link)
    
    print("{0} [{1}]".format(seed,len(unseen)))
    
    #return 새로찾은 URL
    return unseen
```

```python
## 제약 ###########################
threshold = 2

def urlExtractor(seed, limit) :
    if limit > threshold :
        return []

    html = download(seed)
    dom = BeautifulSoup(html.text, "lxml")
    links = [ _["href"] for _ in dom.select("a") if _.has_attr("href")]
    
    unseen = list()
    
    for link in links :
        if len(link) > 1 and link[0] == "/" :
            newLink = requests.compat.urljoin(seed,link)
            if newLink not in unseen :
                unseen.append((newLink,limit+1))
#            else
#                print("Skipped2", link)
        elif link.startswith("http") :
            unseen.append((link,limit+1))
#        else :
#            print("Skipped1", link)
    
#    print("{0} [{1}]".format(seed,len(unseen)))
    
    #return 새로찾은 URL
    return unseen
```

```python
## 네이버 제약 ###########################
threshold = 2

def urlExtractor(seed, limit) :
    if limit > threshold :
        return []

    html = download(seed)
    dom = BeautifulSoup(html.text, "lxml")
    
    #########################################
    iframe = dom.select_one("iframe")
    if not iframe or not iframe.has_attr("src") :
        return []
    
    html = download(requests.compat.urljoin(seed,iframe["src"]))
    dom = BeautifulSoup(html.text,"lxml")
    #########################################
    
    links = [ _["href"] for _ in dom.select("a") if _.has_attr("href")]
    
    unseen = list()
    
    for link in links :
        if len(link) > 1 and link[0] == "/" :
            newLink = requests.compat.urljoin(seed,link)
            if newLink not in unseen :
                unseen.append((newLink,limit+1))
#            else
#                print("Skipped2", link)
        elif link.startswith("http") :
            unseen.append((link,limit+1))
#        else :
#            print("Skipped1", link)
    
#    print("{0} [{1}]".format(seed,len(unseen)))
    
    #return 새로찾은 URL
    return unseen
```

```python
## 뽐뿌 게시판

import requests
# 200 => OK
# 400 => My Fault
# 500 => You Fault => Traffic 과도하게 증가
header={"User-Agent":"Mozilla/5.0 (Windows NT 10.0; WOW64; Trident/7.0; rv:11.0) like Gecko"}

def download(url, params ={}) :
    resp = None
    try :
        resp = requests.get(url, params = params, headers = header)
        resp.raise_for_status()
    except requests.exceptions.HTTPError as e :
        if 500 <= e.response.status_code < 600 and retries > 0:
            print(retries)
            resp = download(url, params, retries - 1)
        else :
            print(e.response.status_code)
            print(e.response.reason)
            print(e.response.headers)
    
    return resp
```

```python
from bs4 import BeautifulSoup

url = "http://www.ppomppu.co.kr/zboard/zboard.php"
params = {"id":"ppomppu"}

html = download(url, params)
dom = BeautifulSoup(html.content,"lxml")
```

```python
itemList = list()

for _ in dom.select("tr[class^='list']")[1:]:
    item = dict()
    tdList = _.find_all("td", recursive=False)
    item["name"] = tdList[3].select_one("a > font").text.strip()
    item["img"] = requests.compat.urljoin(url,tdList[3].select_one("img.thumb_border")["src"])  ##"이미지 경로"
    item["url"] = requests.compat.urljoin(url,tdList[3].select_one("img.thumb_border").find_parent()["href"])## "상품경로"
    item["reg"] = tdList[4].text.strip()
    thumbs = tdList[5].text.strip().split("-")
    item["thumbsup"] = 0 if len(thumbs) < 2 else thumbs[0]
    item["thumbsdown"] = 0 if len(thumbs) < 2 else thumbs[1]
    item["hit"] = tdList[6].text.strip()
    itemList.append(item)

print(len(itemList))
itemList
```
