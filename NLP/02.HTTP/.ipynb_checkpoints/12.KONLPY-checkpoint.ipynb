```python
import os

k = 50
b = 0.5

corpus = ""
base = "./news/"
for _ in [ _ for _ in os.listdir(base) if _.startswith("article")]:
    with open(base + _, encoding="utf-8") as fp :
        corpus += fp.read()
```

```python
import matplotlib.pyplot as plt
        
x = list()
y = list()
_text = ""
# y => unique count

for _ in corpus :
    _text += _
    obj = Text(word_tokenize(_text))
    x.append(k * len(obj) ** b)
    y.append(len(set(obj)))
    
plt.plot(x,y,"r-")
```
