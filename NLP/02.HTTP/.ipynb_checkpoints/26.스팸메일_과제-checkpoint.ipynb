```python
import json

mailList = list()
offset = 0
limit = 15
total = 0
read = True

while read : 
    url = 'https://cmail.daum.net/v2/mails?offset=' + str(offset) + '&limit='+ str(limit) +'&folderId=INBOX&labelIds='
    print(url)
    browser.get(url)
    pre = browser.find_element_by_tag_name("pre").text
    dict = json.loads(pre)
    for mail in dict["mails"] :
        #print(mail)
        mailList.append(mail)

    total = int(dict["total"])
    offset += limit
    if offset > total or total == 0 :
        read = False
        
    #print(total , offset, limit)
    #dict = browser.page_source
    #dict["total"] , dict["mails"]

offset = 0
limit = 15
total = 0
read = True

while read : 
    url = 'https://cmail.daum.net/v2/mails?offset=' + str(offset) + '&limit='+ str(limit) +'&folderId=SPAM&labelIds='
    print(url)
    browser.get(url)
    pre = browser.find_element_by_tag_name("pre").text
    dict = json.loads(pre)
    for mail in dict["mails"] :
        #print(mail)
        mailList.append(mail)

    total = int(dict["total"])
    offset += limit
    if offset > total or total == 0 :
        read = False
        
    #print(total , offset, limit)
    #dict = browser.page_source
    #dict["total"] , dict["mails"]
    
mailList

#browser.get('https://mail.daum.net/#INBOX')
#browser.get('https://mail.daum.net/#SPAM')

# 메일함
# https://cmail.daum.net/v2/mails?offset=0&limit=15&folderId=INBOX&labelIds=
# 스팸
# https://cmail.daum.net/v2/mails?offset=0&limit=15&folderId=SPAM&labelIds=
```

```python
V = list(set([ term for _ in mailList for term in  _["subject"].lower().split()]))
# V = 모든단어
len(V), V
```

```python
def ngram(term, n=2) :
    return [term[i:i+n] for i in range(len(term) - n + 1)]
```

```python
from nltk.tokenize import word_tokenize
from konlpy.tag import Komoran

ma = Komoran()


VSET = set()
for mail in mailList :
    subject = mail['subject']
    tokens2 = word_tokenize(subject) 
    tokens3 = [_ for token in tokens2 for _ in ma.pos(token)]
    tokens4 = [token[0] for token in tokens3] 
    tokens5 = [token[0] for token in tokens3 if token[1].startswith('N')] 
    tokens6 = [_ for token in tokens4 for _ in ngram(token)]
    VSET.update(tokens6)
len(VSET), VSET
```

```python
def termList(mailList) :
    VSET = set()
    for mail in mailList :
        subject = mail['subject']
        tokens2 = word_tokenize(subject) 
        tokens3 = [_ for token in tokens2 for _ in ma.pos(token)]
        tokens4 = [token[0] for token in tokens3] 
        tokens5 = [token[0] for token in tokens3 if token[1].startswith('N')] 
        tokens6 = [_ for token in tokens4 for _ in ngram(token)]
        VSET.update(tokens4)
    #len(VSET), VSET
    return VSET

def termValue(subject) :
    tokens2 = word_tokenize(subject) 
    tokens3 = [_ for token in tokens2 for _ in ma.pos(token)]
    tokens4 = [token[0] for token in tokens3] 
    tokens5 = [token[0] for token in tokens3 if token[1].startswith('N')] 
    tokens6 = [_ for token in tokens4 for _ in ngram(token)]
    return tokens4
```

```python
V = termList(mailList)
V
```

```python
N = len(mailList)
# 문서갯수
N
```

```python
inboxData = [ _ for _ in mailList if _['folderId'] == 'INBOX' ]
spamData = [ _ for _ in mailList if _['folderId'] == 'SPAM' ]
len(inboxData), len(spamData)
```

```python
from collections import defaultdict

Tct = defaultdict(int)
for data in inboxData :
    Nc = len(inboxData)
    PriorC = Nc/N
    
    #for term in data['subject'].lower().split() :
    for term in list(termValue(data['subject'])) :
        Tct[term] += 1

_Tct = defaultdict(int)
for data in spamData :
    _Nc = len(spamData)
    _PriorC = _Nc/N
    
    #for term in data['subject'].lower().split() :
    for term in list(termValue(data['subject'])) :
        _Tct[term] += 1

Tct, _Tct
```

```python
condProbC = defaultdict(float)
_condProbC = defaultdict(float)

countSum = sum(Tct.values())
_countSum = sum(_Tct.values())

for term, freq in Tct.items() :
    condProbC[term] = (freq+1)/(countSum + len(V))
for term, freq in _Tct.items() :
    _condProbC[term] = (freq+1)/(_countSum + len(V))
    
condProbC, _condProbC
```

```python
# P(C) Multi(P(Tcd|C)) ==> log(P(C)) + Sum(P(Tct|C))
from math import log, exp

#==> Joint Prob ==> Conditional Independence
for subject in testingData :

    # Prior Prob
    result = log(PriorC)
    _result = log(_PriorC)
    
    for term in list(termValue(subject)) :
        #==> Multi ==> Log Sum
        result += log((Tct[term]+1)/(countSum+len(V)))
        _result += log((_Tct[term]+1)/(_countSum+len(V)))

    if result > _result : ## Optimal
        print("INBOX", result, exp(result), subject)
    else :
        print("SPAM", _result, exp(_result), subject)   
        
```
