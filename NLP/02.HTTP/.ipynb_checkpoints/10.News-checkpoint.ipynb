```python
import requests
# 200 => OK
# 400 => My Fault
# 500 => You Fault => Traffic 과도하게 증가
header={"User-Agent":"Mozilla/5.0 (Windows NT 10.0; WOW64; Trident/7.0; rv:11.0) like Gecko"}

def download(url, params ={}, retries = 3) :
    resp = None
    try :
        resp = requests.get(url, params = params, headers = header)
        resp.raise_for_status()
    except requests.exceptions.HTTPError as e :
        if 500 <= e.response.status_code < 600 and retries > 0:
            print(retries)
            resp = download(url, params, retries - 1)
        else :
            print(e.response.status_code)
            print(e.response.reason)
            print(e.response.headers)
    
    return resp
```

```python
import json
from bs4 import BeautifulSoup
url = "https://media.daum.net/society/"

html = download(url)
dom = BeautifulSoup(html.text, "lxml")
cnt = 0

#browser.page_source
for _ in dom.select_one(".section_cate").select("a.link_txt"):
    if _["href"].startswith("http") :
        subUrl = _["href"]
        subHtml = download(subUrl)
        subDom = BeautifulSoup(subHtml.text, "lxml")
        title = subDom.select_one(".tit_view").text
        content = subDom.select_one("section").text
        info_view = subDom.select("span.info_view > span.txt_info")

        info_dict = dict()
        
        for _ in info_view :
            if _.text.startswith("입력") :
                info_dict["입력"] = _.text.replace("입력 ","").replace(".","").replace(":","").replace(" ","")
            elif _.text.startswith("수정") :
                info_dict["수정"] = _.text.replace("수정 ","").replace(".","").replace(":","").replace(" ","")
            else :
                info_dict["기자"] = _.text
        
        cnt = cnt+1
        file_name = "./news/article" + str(cnt).rjust(3, '0') + "-" + info_dict["입력"] + ".txt"

        print(title, file_name)
        
        with open(file_name,"w", encoding='utf8') as fp:
            fp.write(title)
            fp.write(str(info_dict))
            fp.write(content)
    
#fp.close()
        
```
